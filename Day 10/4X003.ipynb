#### Q. What will be the standard deviation of the given array? 


```python
w=[1,7,2,3,8,4,11,9,5,6]
```

Hint: 
1.numpy.sum can be used to find the sum of an array example np.sum(w)

2. nupmy.power can be used to find square of any value for example np.power((a),2),

3. np.sqrt can be used to find the squar root of any value example np.sqrt(a)

```python
import numpy as np
pop_std = np.std(w)
sample_std = np.std(w, ddof=1)
print("Population Std Dev:", pop_std)
print("Sample Std Dev:", sample_std)


```
