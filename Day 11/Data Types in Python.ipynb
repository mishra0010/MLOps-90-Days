# Basic Data types in Python.
1. Integer
2. Float
3. complex
4. Boolean
5. String


```python
# Integer
a = 5
print(a)
print(type(a))
```

```python
# Floating Point
b = 3.5
print(b)
type(b)
```

```python
# Complex Numbers
c = 1 + 3j
print(c)
type(c)
```

#Conversion of one data type to another

```python
# Int to float
x = 10
print(x)
type(x)
```

```python
y = float(x)
print(y)
type(y)
```

```python
# Float to int
x = 5.7
print(x)
type(x)
```

```python
y = int(x)
print(y)
type(y)
```

```python
# Boolean
a = True
print(a)
type(a)
```

```python
b = False
print(b)
type(b)
```

```python
a = 7 < 3
print(a)
type(a)
```

```python
a = 7 > 3
print(a)
type(a)
```

```python
# Strings
my_string = "Machine Learning"
print(my_string)
type(my_string)
```

```python
print("Hello"*5)
```
