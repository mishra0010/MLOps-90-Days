# Operators in Python:

1. Airthmetic
2. Assignment
3. Comparison
4. Logical
5. Identity
6. Membership

# Airthmetic Operators

```python
num_1 = 20
num_2 = 10

#addition
sum = num_1 + num_2
print(sum)

#substraction
sub = num_1 - num_2
print(sub)

#Multiplication
prod = num_1 * num_2
print(prod)

#division
div = num_1/num_2
print(div)

#exponent
exp = num_1**num_2
print(exp)

#remainder
rem = num_1%num_2
print(rem)

#int division
int_div = num_1//num_2
print(int_div)
```

# Assignment Operator

```python
a = 5
print(a)
```

```python
a = 5
a += 5 # a = a + 5
print(a)
a -= 3
print(a)
```

# Comparison Operator

```python
a = 5
b = 10

print(a>b)
print(b>a)
print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
```

# Logical operator

```python
a = 10
print(a>20 and a>5)
print(a>20 or a>5)
print(not(a>8 and a>5))
```

# Identity Operator

```python
x = 5
y = 5
print(x is y)
```

```python
x = 5
y = 10
print(x is y)
```

```python
x = 5
y = 10
print(x is not y)
```

# Membership Operator

```python
a = 5
b = 10

c = [1,2,3,4,5]

print(a in c)
print(b in c)
```

```python
a = 5
b = 10

c = [1,2,3,4,5]

print(a not in c)
print(b not in c)
```
