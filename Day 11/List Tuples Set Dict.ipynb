# Types of objects in python
1. Immutable
2. Mutable


#LIST

```python
my_list = [1,2,3,4,5]
print(my_list)
type(my_list)
```

```python
my_list = [2,3,1.8,"Strng",True]
print(my_list)
type(my_list)
```

```python
my_list = [1,2,3,4]
my_list.append(9)
print(my_list)
my_list[0] = 8
print(my_list)
```

```python
# Allows Duplicate
list1 = [1,2,3,4,4]
print(list1)
len(list1)
```

```python
# Empty List
list2 = []
print(list2)
```

```python
list2.append(5)
print(list2)
```

```python
print(list2)
del list2[0]
print(list2)
```

# TUPLE

```python
tuple_1 = (2,3,4,5)
print(tuple_1)
type(tuple_1)
```

```python
tuple_2 = (1,2,3.5,"ML",True)
print(tuple_2)
type(tuple_2)
```

```python
my_list = [1,2,3,4,4]
my_tuple = tuple(my_list)
print(my_tuple)
print(type(my_tuple))
```

```python
my_tuple[1] = 5
```
